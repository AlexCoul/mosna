```python
var_aggreg = mosna.make_features_NAS(
            X=nodes[attributes].astype(float).values, 
            pairs=edges.values, 
            order=1)
```

```python
cluster_params = {
    'reducer_type': 'umap', 
    'n_neighbors': 15, 
    'metric': 'cosine', #'manhattan', # or 'euclidean',
    'min_dist': 0.0,
    'clusterer_type': 'leiden', 
    'dim_clust': 2, 
    'k_cluster': 15, 
    'resolution_parameter': 0.05,
}

cluster_labels, cluster_dir, nb_clust, _ = mosna.get_clusterer(var_aggreg, './', **cluster_params)
```

```python

plt.rcParams['figure.dpi'] = 150
plt.rcParams['figure.figsize'] = [12, 12]

nodes['cluster'] = cluster_labels
sns.scatterplot(data = nodes, x = 'X', y = 'Y', hue = 'cluster', palette="bright")
```
