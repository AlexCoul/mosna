We will extract relevant data from the Phillips data to test several Niche detection methods

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('raw/41467_2021_26974_MOESM3_ESM.xlsx')
```

```python
df
```

```python
df_preprocessed = df.drop([0,1]).reset_index(drop=True)
df_preprocessed.columns = df.iloc[1]
display(df_preprocessed)
display(df_preprocessed.columns)
```

```python
df_one_spot_only = df_preprocessed[df_preprocessed['Spots'] == 1]
display(df_one_spot_only)
```

```python
marker_col_names = ['FOXP3', 'GATA3', 'MUC-1', 'Vimentin', 'T-bet', 'CD62L', 'Cytokeratin', 'PD-L1', 'Ki-67', 'CD15', 'CD30', 'CD2', 'GranzymeB', 'CD5', 'CD39', 'MMP9', 'CD4', 'LAG3', 'CD25', 'CD56', 'CD20', 'PD-1', 'CD11c', 'CLA-CD162', 'IDO-1', 'VISTA', 'HLA-DR', 'ICOS', 'BCL-2', 'CD3', 'CD69', 'CD8', 'CD7', 'CD45RA', 'CD45', 'CD1a', 'CD57', 'B-catenin', 'CD45RO', 'CD71', 'CD34', 'CD68', 'CD38', 'CollagenIV', 'CD31', 'Podoplanin', 'CD138', 'CD163', 'Mastcell-tryptase', 'MMP12', 'CD164', 'p53', 'CCR6', 'CD16', 'CD11b', 'CCR4', 'EGFR', 'HOECHST1:Cyc_1_ch_1', 'DRAQ5']
cols_to_keep = ['ClusterName', 'X', 'Y'] + marker_col_names
```

```python
df_test_selection_spot_1 = df_one_spot_only[cols_to_keep].copy()
df_test_selection_spot_1
```

```python
df_test_selection_spot_1.to_csv('preprocessed/test_selection_spot_1.csv', index=False)
```

Now export as 3 separate files

```python
df_test_coords = df_test_selection_spot_1[['X', 'Y']]
df_test_cell_types = df_test_selection_spot_1['ClusterName']
df_markers = df_test_selection_spot_1.drop(['ClusterName', 'X', 'Y'], axis=1)

df_test_coords.to_csv('preprocessed/test_coords.csv', index=False)
df_test_cell_types.to_csv('preprocessed/test_cell_types.csv', index=False)
df_markers.to_csv('preprocessed/test_markers.csv', index=False)
```

We will now automate the process to obtain the same results for all spots, and not just for spot 1. Therefore, we will loop through all spots in the dataset

```python
df_test_selection_spot_1 = df_one_spot_only[cols_to_keep].copy()
df_test_selection_spot_1

all_spots = df_preprocessed['Spots'].unique()

for spot in all_spots:
    df_spot = df_preprocessed[df_preprocessed['Spots'] == spot]
    df_spot_selection = df_spot[cols_to_keep].copy()
    df_spot_selection.to_csv(f'preprocessed/spot_{spot}.csv', index=False)

    df_spot_coords = df_spot_selection[['X', 'Y']]
    df_spot_cell_types = df_spot_selection['ClusterName']
    df_spot_markers = df_spot_selection.drop(['ClusterName', 'X', 'Y'], axis=1)

    df_spot_coords.to_csv(f'preprocessed/spot_{spot}_coords.csv', index=False)
    df_spot_cell_types.to_csv(f'preprocessed/spot_{spot}_cell_types.csv', index=False)
    df_spot_markers.to_csv(f'preprocessed/spot_{spot}_markers.csv', index=False)
```
